```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('../../../data/kc_house_data_update5.csv')
```

```python
data.columns
```

```python
categoricals = ['bedrooms','bathrooms','floors','waterfront','view','condition','renovated']
```

```python
data_dummies = pd.get_dummies(data[categoricals], prefix = categoricals, columns = categoricals)
```

```python
data_dummies
```

```python
data.drop(categoricals, axis = 1, inplace = True)
data = pd.concat([data, data_dummies], axis = 1)
```

```python
data.to_csv(path_or_buf = '../../../data/kc_house_data_with_dummies.csv', index = False)
```
